```python
# import necessary libraries 
from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np

import logging
logger = logging.getLogger()
logger.setLevel(logging.CRITICAL)
```

# Some images of class 0

```python
# Get the path and images for the train set of class 0 
folder_dir = '../data/train+val/train/0' # REMARK: data must be stored in the right folder, otherwise the path has to be changed
files = os.listdir(folder_dir)
```

```python
# Select 5 random images within this class
a = np.random.randint(len(files), size=5)
```

```python
# Plot the 5 images
fig, ax = plt.subplots(1, 5, figsize=(15, 3))

for n, i in enumerate(a):
    im_dir = files[i]
    im = Image.open(folder_dir + "/" + im_dir)
    ax[n].imshow(im)

fig.suptitle('Class 0: no metastases')
plt.show()
```
