### Deep learning for medical image analysis 

In the coming decade, artificial intelligence will affect many aspects of human life and society ranging from home automation and self-driving cars to algorithmic detection of “fake news” on social media. This technology will also inevitably influence healthcare and in turn medical imaging. One of the most active fields of research in artificial intelligence is deep machine learning [1][2]. This concept works by learning abstract, hierarchical feature representations of input data that are predictive of a certain target (e.g. a class label in the case of classification). In contrast to classical machine learning methods, deep learning eliminates the step of “manual” feature engineering that is a tedious process and sub-optimal compared with learned representations.

In the field of medical image analysis, the use of deep learning methods has already resulted in notable progress for a variety of tasks. Some of the first applications where automatic methods were shown to outperform medical experts have been in histopathology image analysis [3]. 


### Medical image analysis challenges

One of the developments that brought about the success of deep learning methods in computer vision was the availability of large, public annotated image datasets. In medical image analysis, this was mirrored in the form of medical image analysis challenges – friendly competitions in which researchers worldwide evaluate their solutions on the same data with the same criteria, in a blinded manner [4]. Organizing a challenge constitutes collecting and publicly distributing a training dataset that can be used to address a specific clinical task (such as detection of lymph node metastases). The ground truth for the testing is set is retained by the organizers of the challenge, which facilitates fair and independent evaluation and comparison of methods. The organization of medical image analysis challenges has had a tremendous impact on the advancement of the state of the art in the field. In the coming decade, such efforts will continue to push the field forward. 

For the main project work, you will use a dataset of image patches [9] derived from the CAMELYON16 challenge [3]. This Patch-CAMELYON dataset is itself published as a challenge on the Kaggle platform [10], where you will submit the results of your project work for independent evaluation. 

Some other examples of histopathology image analysis challenges organized by our group are AMIDA13 [5] and TUPAC16 [6].

### Detection of lymph node metastases

The presence of metastases in sentinel axillary lymph nodes is an important component of breast cancer staging [7]. This is evaluated by examination of histological preparations of lymph node tissue by pathologists. The figure below gives an example of a metastasis in lymph node tissue. Although manual assessment by a pathologists is the current standard practice, this procedure is notoriously tedious to perform and can result in missed metastases, which in turn results in suboptimal therapy planning for the patients. The move towards digitalization of histopathology labs with the introduction of whole-slide image scanners [8], has opened the possibility of automating some of the tasks performed by pathologists with the use of image analysis methods. The goal is to have faster, more accurate and more objective analysis of the tissue. Tedious and error prone tasks such as detection of metastases are a very good use case for introduction of automated image analysis.

- Answer to **Exercise 1 (What is the clinical utility of evaluating the presence of metastases in sentinel lymph nodes in breast cancer patients? In other words, how is this information used in the clinical decision making process for breast cancer patients?)**:



Breast carcinoma can spread to the surrounding tissues of the breast or other parts of the body by migrating form a primary tumor to axillary lymph nodes. This condition is most accurately diagnosed by histological examination of the lymph nodes by biopsy. Breast cancer cells migrate to one or a few lymph nodes, called sentinel lymph nodes, before involving others, which means that the status of these sentinel lymph nodes may predict the status of the remaining lymph nodes, which is also an important indicator about how far the cancer has spread throughout the body. [11]

So, the presence of metastases in sentinel lymph nodes indicates that the cancer is spreading through the lymph system and the body.

- Answer to **Exercise 2 ("How does the introduction of whole-slide imaging change the typical workflow of a pathology lab?")**
:

Before the introduction of whole-slide imaging, assessment of proliferation of breast cancer tumors was likely to happen through the counting of mitotic figures in H&E (hematoxylin and eosin) stained breast cancer sections performed by a pathologist. This is a subjective procedure that requires the following of a strict protocol, rigorous training and is liable to intra-observer variation. 
The identification of mitoses is difficult because of their different possible appearances and other cellular components that can have similar appearances. In addition to that, this procedure has a low reproducibility, since the counting of mitotic figures is performed in only a small area of the tumor which is chosen in a subjective manner. 
All of this makes mitosis counting a labor-intensive and difficult to reproduce task. 

The introduction of whole-slide imaging changes all of this. The data can now be used by image analysis techniques based on AI that take away most of the subjectivity and labour-intensitivity of manual mitosis counting. [5] The introduction of whole-slide imaging changes the workflow of a pathology lab by releaving the pathologist of the simple tasks which are performed by AI and leaving them with the more difficult tasks. This might result in the pahtologist's work-day becoming more demanding but also even more interesting. The introduction of whole-slide imaging also facilitates dynamic workload balancing or even the reallocation of cases in the someone is sick for example. [12]

Answer to **Exercise 3 (The PatchCamelyon dataset is derived from the CAMELYON16 dataset of whole-slide images. Describe how a neural network classification model trained on small image patches can be applied to larger, whole-slide images with the goal of detecting metastases.)**:

This is achieved by using Stochastic Gradient Descent (SGD). This means that you show the network the input vector for a few examples, then compute the outputs and the errors. Then the average gradient (this gradient vector indicates for each weight by what amount the error would increase/decrease if the weight were increased by a tiny amount) for those examples is calculated and the weights are adjusted accordingly. This process is repeated for many small sets of examples from the training set until the average of the objective function stops decreasing. If this SGD was succesfull, the neural network can now generalize to the whole-slide images and should be able to produce sensible answers on this data set. [1]
